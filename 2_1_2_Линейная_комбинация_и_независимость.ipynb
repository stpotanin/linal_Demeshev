# Свойства определителя

[Как найти определитель матрицы?](https://translated.turbopages.org/proxy_u/en-ru.ru.494ecb63-66b0e7e4-90f36f35-74722d776562/https/www.geeksforgeeks.org/how-to-find-the-determinant-of-a-matrix/)

1. Определитель матрицы и ее собственное транспонирование всегда равны, т.е.

$$
\begin{vmatrix}{a}&b& c \\l&m& n \\x&y& z\end{vmatrix}=\begin{vmatrix}{a}&l& x\\b & m&y \\c& n & z\end{vmatrix}
$$

2. Замена любых двух строк или столбцов матрицы местами изменит знак ее определителя, т.е.

$$
\begin{vmatrix}{a}&b& c \\l&m& n \\x&y&z\end{vmatrix}=-\begin{vmatrix}{a}&b& c \\x&y& z \\l&m& n\end{vmatrix}
$$

3. Определитель матрицы с любыми двумя одинаковыми строками или столбцами равен нулю, т.е.

$$
\begin{vmatrix}{l}&m&n\\l&m&n\\x&y&z\end{vmatrix}=0=\begin{vmatrix}{l}&l&x\\m&m&y\\n&n&z\end{vmatrix}
$$

4. Определитель матрицы с нулями в качестве элементов любой из ее строк или столбцов равен нулю, т.е.

$$
\begin{vmatrix}{0}&0&0\\l & m & n \\x& y & z \end{vmatrix}=0
$$

5. Умножение строки определителя на константу $k$ также увеличит значение определителя в $k$раз, т.е.

$$
\begin{vmatrix}{ka}&kb&kc\\l&m&n\\x&y&z\end{vmatrix}=k\begin{vmatrix}{a}&b&c\\l&m&n\\x&y&z\end{vmatrix}
$$

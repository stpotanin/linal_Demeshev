# Задача 05

Задача 5. Доказать, что каждая из систем векторов

$
E = {(2, 1, 2),(3, −1, 4),(2, 4, 1)}$ и $F = {(−1, 0, 1),(2, 1, 0),(1, 2, −1)}
$

является базисом пространства $R^3$ и найти матрицу перехода от $E$ к $F$.

## Решение

Так как количество векторов в каждой системе совпадает с размерностью пространства, то достаточно проверить линейную независимость векторов в каждой системе.

```python
# делаем матрицу
from sympy import Matrix

E = Matrix([[2, 1, 2], [3, -1, 4], [2, 4, 1]])
F = Matrix([[-1, 0, 1], [2, 1, 0], [1, 2, -1]])
# E.det(), F.det()
# базисность доказана
```

матрица перехода от базиса E к базису F составлена из координат векторов системы векторов F в базисе E, записанных в столбцы

Поскольку $E * P = F$, то $P = E^{-1} * F$

```python
# Находим матрицу перехода от базиса E к базису F
P = E.inv() * F
P
```

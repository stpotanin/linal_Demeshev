Поймать двух зайцев сразу (решить две системц уравнений с одинаковой левой частью)

$$
\begin{bmatrix}
1 & 3 & 3 & | & 10 \\
-1 & 1 & 1 & | & 2 \\
2 & 7 & 1 & | & 17 \\
\end{bmatrix}, \begin{bmatrix}
1 & 3 & 3 & | & 10 \\
-1 & 1 & 1 & | & 6 \\
2 & 7 & 1 & | & 6 \\
\end{bmatrix}
$$

методом Гаусса

# Решение

Решаем систему с двойной правой частью

$$
\begin{bmatrix}
1 & 3 & 3 & | & 10 & 10 \\
-1 & 1 & 1 & | & 2 & 6 \\
2 & 7 & 1 & | & 17 & 6 \\
\end{bmatrix}
$$

Приводим левую часть к единичной матрице, используя гауссовские преобразования:

$$
\begin{bmatrix}
1 & 0 & 0 & | & 1 & -2 \\
0 & 1 & 0 & | & 2 & 1 \\
0 & 0 & 1 & | & 1 & 3 \\
\end{bmatrix}
$$

Ответ:

$$
x = \begin{bmatrix}
1\\
2\\
1\\
\end{bmatrix}, y = \begin{bmatrix}
-2\\
1\\
3\\
\end{bmatrix}
$$

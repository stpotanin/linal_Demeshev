# Задача 01

Линейно зависимы ли векторы в R4?
$$
a_1 = (4,−5, 2, 6),
$$
$$
a_2 = (2,−2, 1, 3),
$$
$$
a_3 = (6,−3, 3, 9),
$$
$$
a_4 = (4,−1, 5, 6).
$$

## Решение

Для проверки линейной зависимости векторов можно использовать следующие шаги:

1. Сформировать матрицу, в которой каждый столбец соответствует одному из векторов.

2. Выполнить операции со строками, чтобы упростить матрицу, затем вычислить определитель.

3. Если результирующая матрица имеет ненулевой определитель, векторы линейно независимы. Если определитель равен нулю, то векторы линейно зависят.

# Свойства определителя

[Как найти определитель матрицы?](https://translated.turbopages.org/proxy_u/en-ru.ru.494ecb63-66b0e7e4-90f36f35-74722d776562/https/www.geeksforgeeks.org/how-to-find-the-determinant-of-a-matrix/)

1. Определитель матрицы и ее собственное транспонирование всегда равны, т.е.

$$
\begin{vmatrix}{a}&b& c \\l&m& n \\x&y& z\end{vmatrix}=\begin{vmatrix}{a}&l& x\\b & m&y \\c& n & z\end{vmatrix}
$$

2. Замена любых двух строк или столбцов матрицы местами изменит знак ее определителя, т.е.

$$
\begin{vmatrix}{a}&b& c \\l&m& n \\x&y&z\end{vmatrix}=-\begin{vmatrix}{a}&b& c \\x&y& z \\l&m& n\end{vmatrix}
$$

3. Определитель матрицы с любыми двумя одинаковыми строками или столбцами равен нулю, т.е.

$$
\begin{vmatrix}{l}&m&n\\l&m&n\\x&y&z\end{vmatrix}=0=\begin{vmatrix}{l}&l&x\\m&m&y\\n&n&z\end{vmatrix}
$$

4. Определитель матрицы с нулями в качестве элементов любой из ее строк или столбцов равен нулю, т.е.

$$
\begin{vmatrix}{0}&0&0\\l & m & n \\x& y & z \end{vmatrix}=0
$$

5. Умножение каждой строки определителя на константу m также увеличит значение определителя в m раз, т.е.

$$
\begin{vmatrix}{ka}&kb&kc\\l&m&n\\x&y&z\end{vmatrix}=k\begin{vmatrix}{a}&b&c\\l&m&n\\x&y&z\end{vmatrix}
$$

```python
# делаем матрицу
from sympy import Matrix

a1 = [4, -5, 2, 6]
a2 = [2, -2, 1, 3]
a3 = [6, -3, 3, 9]
a4 = [4, -1, 5, 6]

a = [a1, a2, a3, a4]

a = Matrix(a)
a
```

```python
# сначала кодом
# Вычисляем определитель
det_a = a.det()
print(det_a)
```

```python
# упрощаем и находим определитель
# меняем строки 0 и 1, 2 и 3, знак определителя не меняется, т.к. 2 замены
a[0, :], a[1, :] = a[1, :], a[0, :]
a[2, :], a[3, :] = a[3, :], a[2, :]
a
```

```python
# вычитаем кратно первую строку из остальных, чтобы обнулить первые элементы строк
a[1, :] = a[1, :] - a[0, :] * 2
a[2, :] = a[2, :] - a[0, :] * 2
a[3, :] = a[3, :] - a[0, :] * 3
a
```

```python
# вычитаем кратно вторую строку из остальных ниже, чтобы обнулить вторыые элементы строк
a[2, :] = a[2, :] - a[1, :] * (-3)
a[3, :] = a[3, :] - a[1, :] * (-3)
a
# по свойству 4 определитель равен 0, векторы линейно зависят
```

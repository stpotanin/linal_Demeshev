# Задача 6

Проверить, образуют ли подпространство векторы пространства $R^n$ , координаты которых удовлетворяют уравнению $x1 + x2 + . . . + xn = a$, где $a ∈ R$ — фиксированное число.

## Решение

Чтобы определить, образуют ли векторы пространства $ \mathbb{R}^n $, координаты которых удовлетворяют уравнению $ x_1 + x_2 + \dots + x_n = a $, подпространство, нужно проверить выполнение трёх условий:

1. **Наличие нулевого вектора**: Подпространство должно содержать нулевой вектор.
2. **Замкнутость относительно сложения**: Если $ \mathbf{u} $ и $ \mathbf{v} $ принадлежат подпространству, то их сумма $ \mathbf{u} + \mathbf{v} $ также должна принадлежать подпространству.
3. **Замкнутость относительно умножения на скаляр**: Если $ \mathbf{u} $ принадлежит подпространству и $ c $ — любое действительное число, то $ c\mathbf{u} $ также должно принадлежать подпространству.

Теперь проверим выполнение этих условий для данного множества векторов:

### 1. Наличие нулевого вектора
Нулевой вектор в пространстве $ \mathbb{R}^n $ имеет вид $ \mathbf{0} = (0, 0, \dots, 0) $. Подставим этот вектор в уравнение $ x_1 + x_2 + \dots + x_n = a $:

$$
0 + 0 + \dots + 0 = 0
$$

Так как $ a $ фиксировано и не обязательно равно нулю, нулевой вектор не будет удовлетворять данному уравнению, если $ a \neq 0 $.

Таким образом, множество векторов не содержит нулевой вектор, если $ a \neq 0 $, что означает, что это множество не может быть подпространством.

### 2. Замкнутость относительно сложения
Предположим, что $ \mathbf{u} = (u_1, u_2, \dots, u_n) $ и $ \mathbf{v} = (v_1, v_2, \dots, v_n) $ — два вектора, которые принадлежат данному множеству. Тогда они удовлетворяют уравнению:

$$
u_1 + u_2 + \dots + u_n = a \quad \text{и} \quad v_1 + v_2 + \dots + v_n = a
$$

Рассмотрим их сумму $ \mathbf{w} = \mathbf{u} + \mathbf{v} = (u_1 + v_1, u_2 + v_2, \dots, u_n + v_n) $. Тогда:

$$
(u_1 + v_1) + (u_2 + v_2) + \dots + (u_n + v_n) = (u_1 + u_2 + \dots + u_n) + (v_1 + v_2 + \dots + v_n) = a + a = 2a
$$

Таким образом, если $ a \neq 0 $, то сумма двух векторов $ \mathbf{u} $ и $ \mathbf{v} $ не принадлежит множеству, потому что сумма их координат равна $ 2a $, а не $ a $. Следовательно, множество не замкнуто относительно сложения.

### 3. Замкнутость относительно умножения на скаляр
Рассмотрим вектор $ \mathbf{u} = (u_1, u_2, \dots, u_n) $, который принадлежит данному множеству, то есть $ u_1 + u_2 + \dots + u_n = a $. Если умножить его на скаляр $ c $, получим:

$$
c\mathbf{u} = (cu_1, cu_2, \dots, cu_n)
$$

Сумма координат нового вектора будет равна:

$$
cu_1 + cu_2 + \dots + cu_n = c(u_1 + u_2 + \dots + u_n) = ca
$$

Если $ c \neq 1 $, то это выражение не равно $ a $, следовательно, вектор $ c\mathbf{u} $ не принадлежит множеству, если $ c \neq 1 $. Это означает, что множество не замкнуто относительно умножения на скаляр.

### Заключение
Множество векторов, координаты которых удовлетворяют уравнению $ x_1 + x_2 + \dots + x_n = a $, не является подпространством пространства $ \mathbb{R}^n $, если $ a \neq 0 $.

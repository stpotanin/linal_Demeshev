### Найти определитель матрицы 3 x 3 методом перестановок

$$
$$
\begin{bmatrix}
1 & 4 & 2 \\
0 & 2 & 0 \\
3 & 9 & 5 \\
\end{bmatrix}
$$
$$

# Решение

```python
import math

def determinant(matrix):
  """Вычисляет определитель матрицы с помощью перестановок.
  """
  n = len(matrix)
  if n == 1:
    return matrix[0][0]
  elif n == 2:
    return matrix[0][0] * matrix[1][1] - matrix[0][1] * matrix[1][0]
  else:
    determinant_value = 0
    for j in range(n):
      submatrix = [row[:j] + row[j+1:] for row in matrix[1:]]
      sign = (-1)**j
      determinant_value += sign * matrix[0][j] * determinant(submatrix)
    return determinant_value

# Пример использования
matrix = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
det = determinant(matrix)
print(f"Определитель матрицы: {det}")
```

Найти обратную матрицу $B$ для

$$
A = \begin{bmatrix}
3 & 2 & 4 \\
4 & 3 & 6 \\
6 & 4 & 9 \\
\end{bmatrix}
$$

методом Гаусса

# Решение

Нам нужно решить систему $A \cdot B = I$ ,

фактически решить три системы, записанные в виде единичной матрицы.

Приписываем правую часть в виде единичной матрицы:

$$
\begin{bmatrix}
3 & 2 & 4 & | & 1 & 0 & 0 \\
4 & 3 & 6 & | & 0 & 1 & 0 \\
6 & 4 & 9 & | & 0 & 0 & 1 \\
\end{bmatrix}
$$

Если $det(A) \neq 0$, то, с помощью гаусовкких преобразований получаем единичную матрицу слева:

$$
\begin{bmatrix}
1 & 0 & 0 & | & 3 & -2 & 0 \\
0 & 1 & 0 & | & 0 & 3 & -2 \\
0 & 0 & 1 & | & -2 & 0 & 1 \\
\end{bmatrix}
$$

Ответ:

$$
A^{-1} = \begin{bmatrix}
3 & -2 & 0 \\
0 & 3 & -2 \\
-2 & 0 & 1 \\
\end{bmatrix}
$$

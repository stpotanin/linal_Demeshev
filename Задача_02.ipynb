## Вариант 1
Даны два вектора $a=(7,8,9), b=(6,3,5)$. Найдите $\cos \left(a, b\right)$. Округлите полученный ответ с точностью до двух знаков после запятой.

## Решение

$\cos \left(a, b\right) = \dfrac{111}{\sqrt{194}\cdot\sqrt{70}} $.

```python
#@title Ответ
from numpy import dot
from numpy.linalg import norm

a = [7, 8, 9]
b = [6, 3, 5]

cos_ab = dot(a, b) / (norm(a) * norm(b))
print(cos_ab)
# (0.93, 0.98) true
```

## Вариант 2
Даны два вектора $a=(7,8,9), b=(6,3,5)$. Найдите $\langle a+2b, b-a \rangle$.

## Решение

$a+2b = (19, 14, 19)$, $b-a = (-1, -5, -4)$;

$\langle a+2b, b-a \rangle = -19 - 5\cdot14 - 4\cdot19 = -165 $.

```python
#@title Ответ
import numpy as np

aa = np.array([7, 8, 9])
bb = np.array([6, 3, 5])
print(np.dot(aa + 2 * bb, bb - aa))

# -165 true
```

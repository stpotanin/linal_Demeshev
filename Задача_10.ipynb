## Вариант 1
Пусть $A: \begin{pmatrix}v_{1} \\v_{2}\end{pmatrix} \to \begin{pmatrix}1/{2} \cdot v_{1} + \sqrt{3}/2 \cdot v_{2} \\-\sqrt{3}/2 \cdot v_{1} + 1/{2} \cdot v_{2} \end{pmatrix} $. Выберите верные утверждения:

* Операция $A$ отражает вектор $v$ по горизонтальной оси с последующим вращением на 60 градусов по часовой стрелке.

* Операция $A$ поворачивает вектор $v$ на 60 градусов против часовой стрелки.

* Операция $A$ поворачивает вектор $v$ на 60 градусов против часовой стрелки с последующим отражением по горизонтальной оси.

* Операция $A$ поворачивает вектор $v$ на 60 градусов по часовой стрелке.

## Решение

* Операция $A$ отражает вектор $v$ по горизонтальной оси с последующим вращением на 60 градусов по часовой стрелке. false

* Операция $A$ поворачивает вектор $v$ на 60 градусов против часовой стрелки. false

* Операция $A$ поворачивает вектор $v$ на 60 градусов против часовой стрелки с последующим отражением по горизонтальной оси. false

* Операция $A$ поворачивает вектор $v$ на 60 градусов по часовой стрелке. true

```python
# Посмотрим, что будет происходить
import numpy as np

v1 = [1, 0]             # 0 градусов в полярной
v2 = [1/2, 3**.5/2]     # 30
v3 = [3**.5/2, 1/2]     # 60
v4 = [0, 1]             # 90
A = [[1/2, 3**.5/2],
     [-3**.5/2, 1/2]]
# проверка
print([round(x, 3) for x in np.dot(A, v1)])
print([round(x, 3) for x in np.dot(A, v2)])
print([round(x, 3) for x in np.dot(A, v3)])
print([round(x, 3) for x in np.dot(A, v4)])
# вывод: Операция  A  поворачивает вектор  v  на 60 градусов по часовой стрелке
```

## Вариант 2
Пусть $A: \begin{pmatrix}v_{1} \\v_{2}\end{pmatrix} \to \begin{pmatrix}1/\sqrt{2} \cdot v_{1} + 1/\sqrt{2} \cdot v_{2} \\1/\sqrt{2} \cdot v_{1} - 1/\sqrt{2} \cdot v_{2} \end{pmatrix} $. Выберите верные утверждения:

* Операция $A$ отражает вектор $v$ по горизонтальной оси с последующим вращением на 45 градусов по часовой стрелке.

* Операция $A$ отражает вектор $v$ по горизонтальной оси с последующим вращением на 45 градусов против часовой стрелки.

* Операция $A$ поворачивает вектор $v$ на 45 градусов против часовой стрелки с последующим отражением по горизонтальной оси.

* Операция $A$ поворачивает вектор $v$ на 45 градусов по часовой стрелке с последующим отражением по горизонтальной оси.

## Решение

* Операция $A$ отражает вектор $v$ по горизонтальной оси с последующим вращением на 45 градусов по часовой стрелке. false

* Операция $A$ отражает вектор $v$ по горизонтальной оси с последующим вращением на 45 градусов против часовой стрелки. false

* Операция $A$ поворачивает вектор $v$ на 45 градусов против часовой стрелки с последующим отражением по горизонтальной оси. false

* Операция $A$ поворачивает вектор $v$ на 45 градусов по часовой стрелке с последующим отражением по горизонтальной оси. true

```python
# Посмотрим, что будет происходить
import numpy as np

v1 = [1, 0]                 # 0 градусов в полярной
v2 = [1/2**.5, 1/2**.5]     # 45
v3 = [0, 1]                 # 90
v4 = [-1/2**.5, 1/2**.5]    # 135

A = [[1/2**.5, 1/2**.5],
     [1/2**.5, -1/2**.5]]
# проверка
print([round(x, 3) for x in np.dot(A, v1)])
print([round(x, 3) for x in np.dot(A, v2)])
print([round(x, 3) for x in np.dot(A, v3)])
print([round(x, 3) for x in np.dot(A, v4)])
# вывод: Операция  A  поворачивает вектор  v  на 45 градусов по часовой стрелке с последующим отражением по горизонтальной оси
```
